A system of $m$ linear equations with $n$ unknowns ($x_1, ..., x_n$) is written as:

$$\begin{aligned}
a_{11}x_1 + \dots + a_{1n}x_n &= b_1 \\
&\vdots \\
a_{m1}x_1 + \dots + a_{mn}x_n &= b_m
\end{aligned}$$

where $a_{ij}$ are coefficients and $b_i$ are constants2.

To solve these systems efficiently, we use compact matrix notation $Ax = b$:
- Coefficient Matrix ($A$): Contains all $a_{ij}$ values.
- Unknown Vector ($x$): The variables we solve for.
- Constant Vector ($b$): The right-hand side values.

Augmented Matrix: $[A | b]$ is used for manual calculation (Gaussian elimination).

A real-valued system of linear equations always has one of three outcomes:

- Exactly one solution: The lines/planes intersect at a single point.

- Infinitely many solutions: Occurs when there is redundancy (e.g., equations describing the same line).

- No solution: Occurs when equations are contradictory (e.g., parallel lines that never meet).
