A system $Ax = b$ is solved by finding two components:

**Particular Solution** ($x_p$): Any single specific vector that satisfies $Ax = b$.

Usually, we find it by performing Gaussian elimination and setting all the "free variables" to zero. This gives us the simplest possible point that works.

It acts as an "anchor." While the homogeneous solution describes a shape, the particular solution describes where that shape is located in space.

**Homogeneous Solution** ($x_h$): All solutions to the "homogeneous" equation $Ax = 0$.

It represents all vectors that the matrix $\mathbf{A}$ "squashes" into the origin. 

It tells you how much redundancy is in your system. If $\mathbf{x}_h$ has many vectors (a high-dimension null space), it means there are many ways to change your variables without changing the output.

The set of all homogeneous solutions always forms a subspace (like a line or a plane) that passes through the origin. And $x_p$ is just one point in space.

**General Solution**: The set of all solutions, expressed as $x = x_p + x_h$.

You can think of it as taking the "shape" of the homogeneous solution and "shifting" it by the particular solution.

*The Analogy*:

Homogeneous Solution: Imagine a flat floor (a plane) passing through the center of a room (the origin).

Particular Solution: Imagine a pole that is 5 feet tall.

General Solution: If you put that flat floor on top of the 5-foot pole, the floor is still flat, but it is now "floating" 5 feet above the origin. It no longer passes through $(0,0,0)$.

**Example**

$$\begin{bmatrix} 1 & 0 & 8 & -4 \\ 0 & 1 & 2 & 12 \end{bmatrix} \begin{bmatrix} x_1 \\ x_2 \\ x_3 \\ x_4 \end{bmatrix} = \begin{bmatrix} 42 \\ 8 \end{bmatrix}$$

**Homogeneous Solution**

For $x_3$, the column is $\begin{bmatrix} 8 \\ 2 \end{bmatrix}$. We can cancel it using $-8$ of column 1 and $-2$ of column 2. This gives vector $h_1 = [8, 2, -1, 0]^\top$.

For $x_4$, the column is $\begin{bmatrix} -4 \\ 12 \end{bmatrix}$. We can cancel it using $4$ of column 1 and $-12$ of column 2. This gives vector $h_2 = [-4, 12, 0, -1]^\top$.


**Particular Solution** ($x_p$)

Set the free variables ($x_3, x_4$) to 0. We can immediately see:
$x_1 = 42, x_2 = 8$. Thus, $x_p = [42, 8, 0, 0]^\top$.

**General Solution**:

$$x = \begin{bmatrix} 42 \\ 8 \\ 0 \\ 0 \end{bmatrix} + \lambda_1 \begin{bmatrix} 8 \\ 2 \\ -1 \\ 0 \end{bmatrix} + \lambda_2 \begin{bmatrix} -4 \\ 12 \\ 0 \\ -1 \end{bmatrix}$$

**Gaussian Elimination**

Gaussian Elimination is an algorithm that uses Elementary Row Operations to transform an augmented matrix into a simpler form where solutions can be easily read.
The Three Elementary Operations:

1. Swapping: Exchange two rows.
2. Scaling: Multiply a row by a non-zero scalar $\lambda$.
3. Pivoting: Add a multiple of one row to another.

**Example**

System:
$$\begin{aligned} x + 2y + z &= 8 \\ 2x + 6y + z &= 15 \\ x + y + 4z &= 12 \end{aligned}$$

Step-by-Step Transformation:

Augment: $\left[ \begin{array}{ccc|c} 1 & 2 & 1 & 8 \\ 2 & 6 & 1 & 15 \\ 1 & 1 & 4 & 12 \end{array} \right]$

Eliminate under first pivot ($x$): $R_2 - 2R_1 \rightarrow R_2$ and $R_3 - R_1 \rightarrow R_3$.$$\left[ \begin{array}{ccc|c} 1 & 2 & 1 & 8 \\ 0 & 2 & -1 & -1 \\ 0 & -1 & 3 & 4 \end{array} \right]$$Eliminate under second pivot ($y$): $R_3 + 0.5R_2 \rightarrow R_3$.$$\left[ \begin{array}{ccc|c} 1 & 2 & 1 & 8 \\ 0 & 2 & -1 & -1 \\ 0 & 0 & 2.5 & 3.5 \end{array} \right] \rightarrow \text{This is REF}$$Back-substitution: $2.5z = 3.5 \implies z = 1.4$. Plug $z$ into $R_2$ to find $y$, then both into $R_1$ to find $x$.


System:

$$\begin{aligned} x + y + z &= 3 \\ x + 2y &= 2 \end{aligned}$$

After RREF:

$$\left[ \begin{array}{ccc|c} 1 & 0 & 2 & 4 \\ 0 & 1 & -1 & -1 \end{array} \right]$$

Pivots: Columns 1 and 2 ($x, y$).Free Variable: $z$.Solution: $x = 4 - 2z$ and $y = -1 + z$.

The solution set is a line in 3D space.


System:

$$\begin{aligned} x + y + z &= 2 \\ x + y + z &= 5 \end{aligned}$$

Transformation:$R_2 - R_1 \rightarrow R_2$:$$\left[ \begin{array}{ccc|c} 1 & 1 & 1 & 2 \\ 0 & 0 & 0 & 3 \end{array} \right]$$

Look at the last row: $0x + 0y + 0z = 3$, or $0 = 3$.This is a mathematical impossibility. Therefore, the system is inconsistent and has no solution. Geometrically, these represent parallel planes that never intersect.


**Minus-1 Trick**

Given Matrix in RREF:

$$A = \begin{bmatrix} 1 & 2 & 0 & 3 \\ 0 & 0 & 1 & 4 \end{bmatrix}$$

Identify missing pivots: Diagonals $(1,1)$ and $(2,2)$ should have pivots. Column 2 and 4 are missing pivots.Augment with -1 rows: Insert rows to make the matrix square, putting a $-1$ where the pivot is missing.

$$\tilde{A} = \begin{bmatrix} 1 & 2 & 0 & 3 \\ 0 & \mathbf{-1} & 0 & 0 \\ 0 & 0 & 1 & 4 \\ 0 & 0 & 0 & \mathbf{-1} \end{bmatrix}$$

The kernel basis is simply the columns where you placed the $-1$:

$$\text{Kernel Basis} = \left\{ \begin{bmatrix} 2 \\ -1 \\ 0 \\ 0 \end{bmatrix}, \begin{bmatrix} 3 \\ 0 \\ 4 \\ -1 \end{bmatrix} \right\}$$
