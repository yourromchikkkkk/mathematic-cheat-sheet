### Linear Dependence

A vector $v$ is a linear combination of a set of vectors $\{x_1, \dots, x_k\}$ if it can be written as:
$$v = \lambda_1x_1 + \dots + \lambda_kx_k = \sum_{i=1}^{k} \lambda_ix_i$$ 

where $\lambda_1, \dots, \lambda_k \in \mathbb{R}$ are scalars. 

Trivial Case: The 0-vector is always a linear combination of any set of vectors by setting all $\lambda_i = 0$.

A set of vectors $\{x_1, \dots, x_k\}$ is:
- Linearly Dependent: If there exists a non-trivial linear combination that equals the zero vector (i.e., at least one $\lambda_i \neq 0$ such that $\sum \lambda_ix_i = 0$).
- Linearly Independent: If the only linear combination that results in the zero vector is the trivial one ($\lambda_1 = \dots = \lambda_k = 0$).

**Hints**

- If a set contains the 0-vector, it is automatically linearly dependent.
- If two vectors in a set are the same, the set is linearly dependent.
- If one vector is a scalar multiple of another (e.g., $x_i = \lambda x_j$), the set is linearly dependent.
- In a vector space $V$, any $m$ linear combinations of $k$ vectors are linearly dependent if $m > k$13.

**Testing for Linear (In)Dependence**

1. Write the vectors as columns of a matrix $A$.
2. Perform row operations to reach Row-Echelon Form (REF).
3. Result: * The vectors are linearly independent if and only if every column is a pivot column.
    - If there is at least one non-pivot column, the set is linearly dependent.

**Explanation of pivot columns**

Note 1:

A pivot column means:
    
    “This column introduces a new independent direction that was not already created by the columns to its left.”

So, each pivot column adds new information. Each non-pivot column adds **NO** new information

Note 2:

When you build a matrix from column vectors. You are implicitly **ordering** the vectors from left to right.


**Axiom 1**

The pivot columns indicate the vectors which are linearly independent of the vectors on the left.


**Axiom 2**

The non-pivot columns can be expressed as linear combinations of the pivot columns on their left.

*Means*

Every non-pivot column is a linear combination of the pivot columns that appear before it.
