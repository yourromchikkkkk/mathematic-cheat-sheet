Matrix: A real-valued $(m, n)$ matrix $\mathbf{A}$ is a rectangular array with $m$ rows and $n$ columns.

$\mathbf{A} \in \mathbb{R}^{m \times n}$ denotes the set of all real-valued $m \times n$ matrices3.

Special Vectors:
- Row Vector: A $(1, n)$ matrix4.
- Column Vector: An $(m, 1)$ matrix5.
- Identity Matrix ($\mathbf{I}_n$): A square $n \times n$ matrix with 1s on the main diagonal and 0s elsewhere.
- Symmetric Matrix: A square matrix where $\mathbf{A} = \mathbf{A}^\top$

The sum of two matrices $\mathbf{A}, \mathbf{B} \in \mathbb{R}^{m \times n}$ is computed element-wise: $(\mathbf{A} + \mathbf{B})_{ij} = a_{ij} + b_{ij}$

Multiplying a matrix $\mathbf{A}$ by a scalar $\lambda$ scales every element: $(\lambda\mathbf{A})_{ij} = \lambda a_{ij}$

For $\mathbf{A} \in \mathbb{R}^{m \times n}$ and $\mathbf{B} \in \mathbb{R}^{n \times k}$, the product $\mathbf{C} = \mathbf{AB} \in \mathbb{R}^{m \times k}$ has elements $c_{ij} = \sum_{l=1}^{n} a_{il}b_{lj}$

**Remember**: $\mathbf{AB} \neq \mathbf{BA}$.

Formed by swapping rows and columns ($b_{ij} = a_{ji}$).

$(\mathbf{AB})^\top = \mathbf{B}^\top\mathbf{A}^\top$

For a square matrix $\mathbf{A}$, the inverse satisfies $\mathbf{AA}^{-1} = \mathbf{I} = \mathbf{A}^{-1}\mathbf{A}$.

If it exists, $\mathbf{A}$ is regular (invertible/nonsingular); otherwise, it is singular.

**Remember**: $(\mathbf{AB})^{-1} = \mathbf{B}^{-1}\mathbf{A}^{-1}$

**Remember**

Associativity: $(\mathbf{AB})\mathbf{C} = \mathbf{A}(\mathbf{BC})$.

Distributivity: $\mathbf{A}(\mathbf{B} + \mathbf{C}) = \mathbf{AB} + \mathbf{AC}$ and $(\mathbf{A} + \mathbf{B})\mathbf{C} = \mathbf{AC} + \mathbf{BC}$.

**REF (Row-Echelon Form)**

1. All-zero rows: Any rows consisting entirely of zeros must be at the very bottom.
2. Pivots: In every non-zero row, the first non-zero number (called the leading entry or pivot) must be to the right of the pivot in the row above it.
3. Zeros below pivots: All entries in a column below a pivot must be zeros.

Example of REF:

$$\begin{bmatrix} 
\mathbf{2} & 4 & 1 \\ 
0 & \mathbf{5} & 3 \\ 
0 & 0 & 0 
\end{bmatrix}$$

(Here, 2 and 5 are the pivots. Notice how they form a staircase descending to the right.)

**RREF (Reduced Row-Enchelon Form)**

A matrix is in RREF if it meets all the requirements of REF, plus two extra strict rules that make the solution to a system of equations immediately obvious.

The Extra Rules for RREF:

1. Pivots must be 1: Every leading entry (pivot) must be exactly $1$.
2. Zeros above pivots: Every pivot must be the only non-zero entry in its column. This means there are zeros both below and above every pivot.

Example of RREF:

$$\begin{bmatrix} 
\mathbf{1} & 0 & 5 \\ 
0 & \mathbf{1} & 2 \\ 
0 & 0 & 0 
\end{bmatrix}$$

(The pivots are now 1, and the columns containing them have no other non-zero numbers. This tells us directly that $x_1 = 5$ and $x_2 = 2$.)
